```python
import pandas as pd
import numpy as np
```

```python
CUBIC_YARD_FT3 = 27       # cubic feet in one cubic yard
UNIT_WEIGHT_WATER = 62.4  # lb/ft^3 (unit weight of water)
```

```python
# This dictionary maps the Excel-style letter codes to descriptive names
mix_inputs_guide = {
    "cement_A": "Weight of cement (lb/yd³)",
    "fly_ash_B": "Weight of fly ash (lb/yd³)",
    "silica_fume_C": "Weight of silica fume (lb/yd³)",
    "other_scm_D": "Weight of other SCM (lb/yd³)",
    "wc_ratio_E": "Target water-cement ratio",
    "air_content_F": "Target air content (%)",
    "percent_fine_G": "Target percent fine aggregate (%)",
    "percent_coarse_H": "Target percent coarse aggregate (%)",
    "percent_other_I": "Target percent other aggregate (%)",
    "sg_cement_J": "Specific gravity of cement",
    "sg_fly_ash_K": "Specific gravity of fly ash",
    "sg_silica_fume_L": "Specific gravity of silica fume",
    "sg_other_scm_M": "Specific gravity of other SCM",
    "sg_fine_N": "Specific gravity of fine aggregate",
    "sg_coarse_O": "Specific gravity of coarse aggregate",
    "sg_other_P": "Specific gravity of other aggregate"
}
# Simple loop to print
for code, description in mix_inputs_guide.items():
    print(code, ":", description)
```

```python
# Functions Q R S T U V W X Y Z AA defined
def calculate_water_weight_Q(cement_A, fly_ash_B, silica_fume_C, other_scm_D, wc_ratio_E):
    total_cementitious = cement_A + fly_ash_B + silica_fume_C + other_scm_D
    return total_cementitious * wc_ratio_E

def calculate_volume_R(cement_A, sg_cement_J):
    return cement_A / (sg_cement_J * UNIT_WEIGHT_WATER)

def calculate_volume_S(fly_ash_B, sg_fly_ash_K):
    return fly_ash_B / (sg_fly_ash_K * UNIT_WEIGHT_WATER)

def calculate_volume_T(silica_fume_C, sg_silica_fume_L):
    return silica_fume_C / (sg_silica_fume_L * UNIT_WEIGHT_WATER)

def calculate_volume_U(other_scm_D, sg_other_scm_M):
    return other_scm_D / (sg_other_scm_M * UNIT_WEIGHT_WATER)

def calculate_air_volume_V(air_content_F):
    return (air_content_F / 100) * CUBIC_YARD_FT3

def calculate_water_volume_W(water_weight_Q):
    return water_weight_Q / UNIT_WEIGHT_WATER

def calculate_total_aggregate_volume_X(volume_R, volume_S, volume_T, volume_U, volume_V, volume_W):
    return (CUBIC_YARD_FT3 - volume_R - volume_S - volume_T - volume_U - volume_V - volume_W)

def calculate_fine_aggregate_Y(percent_fine_G, sg_fine_N, volume_X):
    return UNIT_WEIGHT_WATER * (percent_fine_G / 100) * sg_fine_N * volume_X

def calculate_coarse_aggregate_Z(percent_coarse_H, sg_coarse_O, volume_X):
    return UNIT_WEIGHT_WATER * (percent_coarse_H / 100) * sg_coarse_O * volume_X

def calculate_other_aggregate_AA(percent_other_I, sg_other_P, volume_X):
    return UNIT_WEIGHT_WATER * (percent_other_I / 100) * sg_other_P * volume_X
```

```python
# Sequential User Prompts
project_no = int(input("Enter project number: "))
concrete_class = input("Enter class of concrete: ")

# Weights
cement_A = float(input("Enter cement weight A (lb/yd³): "))
fly_ash_B = float(input("Enter fly ash weight B (lb/yd³): "))
silica_fume_C = float(input("Enter silica fume weight C (lb/yd³): "))
other_scm_D = float(input("Enter other SCM weight D (lb/yd³): "))

# Parameters
wc_ratio_E = float(input("Enter target water-cement ratio E: "))
air_content_F = float(input("Enter target air content F (%): "))

# Aggregate %
percent_fine_G = float(input("Enter percent fine aggregate G (%): "))
percent_coarse_H = float(input("Enter percent coarse aggregate H (%): "))
percent_other_I = float(input("Enter percent other aggregate I (%): "))

# Specific Gravities
sg_cement_J = float(input("Enter specific gravity of cement J: "))
sg_fly_ash_K = float(input("Enter specific gravity of fly ash K: "))
sg_silica_fume_L = float(input("Enter specific gravity of silica fume L: "))
sg_other_scm_M = float(input("Enter specific gravity of other SCM M: "))
sg_fine_N = float(input("Enter specific gravity of fine aggregate N: "))
sg_coarse_O = float(input("Enter specific gravity of coarse aggregate O: "))
sg_other_P = float(input("Enter specific gravity of other aggregate P: "))

# Main Logic
# 1. Weights
Q = calculate_water_weight_Q(cement_A, fly_ash_B, silica_fume_C, other_scm_D, wc_ratio_E)

# 2. Volumes
R = calculate_volume_R(cement_A, sg_cement_J)
S = calculate_volume_S(fly_ash_B, sg_fly_ash_K)
T = calculate_volume_T(silica_fume_C, sg_silica_fume_L)
U = calculate_volume_U(other_scm_D, sg_other_scm_M)
V = calculate_air_volume_V(air_content_F)
W = calculate_water_volume_W(Q)

# 3. Aggregate Calcs
X = calculate_total_aggregate_volume_X(R, S, T, U, V, W)
Y = calculate_fine_aggregate_Y(percent_fine_G, sg_fine_N, X)
Z = calculate_coarse_aggregate_Z(percent_coarse_H, sg_coarse_O, X)
AA = calculate_other_aggregate_AA(percent_other_I, sg_other_P, X)

# Weight Chart
print("\n---------------------------------------------")
print(" NDOT Concrete Mix Design – Weight Summary")
print("         (1 Cubic Yard of Concrete)")

print("---------------------------------------------")

# Project and mix details
print(f"Project Number:       {project_no}")
print(f"Class of Concrete:    {concrete_class}")

print("---------------------------------------------")

# Cementitious material outputs
print(f"Cement (A):            {cement_A:8.1f} lb")
print(f"Fly Ash (B):           {fly_ash_B:8.1f} lb")
print(f"Silica Fume (C):       {silica_fume_C:8.1f} lb")
print(f"Other SCM (D):         {other_scm_D:8.1f} lb")

print("---------------------------------------------")

# Aggregate outputs
print(f"Fine Aggregate (Y):    {Y:8.0f} lb")
print(f"Coarse Aggregate (Z):  {Z:8.0f} lb")
print(f"Other Aggregate (AA):  {AA:8.0f} lb")

print("---------------------------------------------")

# Water output
print(f"Water (Q):             {Q:8.0f} lb")

# End it with a note
print("---------------------------------------------")
print("End of Mix Design Summary")
```
